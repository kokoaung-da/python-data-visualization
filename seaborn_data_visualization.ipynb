# Seaborn Data Visualization Practice

## Importing required libraries

```python
import pandas as pd
import seaborn as sns
sns.set_style("whitegrid") # Nice clean look: 'white', 'dark', whitegrid', 'darkgrid'
sns.set_palette("Set2") # Change color palette
```

## Loading Dataset

```python
sns.get_dataset_names()
```

```python
tips = sns.load_dataset("tips")
tips.head()
```

## Distribution & Density

### Histogram & KDE
- For understanding value distribution

```python
sns.histplot(data=tips, x="total_bill", bins=20, kde=True)
```

```python
sns.histplot(data=tips, x="total_bill", bins=20, kde=True, hue="sex")
```

```python
sns.boxplot(data=tips, x="total_bill")
```

```python
sns.violinplot(data=tips, x="total_bill")
```

## Relationship Plots

```python
sns.scatterplot(data=tips, x="total_bill", y="tip")
```

```python
sns.scatterplot(data=tips, x="total_bill", y="tip", size="size")
```

```python
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="sex", size="size")
```

```python
dowjones=sns.load_dataset("dowjones")
dowjones.head()
```

```python
sns.lineplot(data=dowjones, x="Date", y="Price")
```

## Categorical Plots

```python
sns.barplot(data=tips, x="day", y="tip", estimator="mean", errorbar=None)
```

```python
titanic=sns.load_dataset("titanic")
titanic.head()
```

```python
sns.countplot(data=titanic, x="class")
```

```python
sns.heatmap(tips.corr(numeric_only=True), annot=True, cmap="coolwarm", linewidths=0.5)
```

## Pairwise

```python
sns.pairplot(data=tips, hue="sex", diag_kind="kde")
```
