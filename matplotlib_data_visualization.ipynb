# Matplotlib Data Visualization

```python
import warnings
warnings.filterwarnings('ignore')
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sns.get_dataset_names()
```

## Line Plot

```python
dowjones=sns.load_dataset("dowjones")
dowjones
```

```python
fig, ax= plt.subplots(figsize=(8,6))
ax.plot(dowjones["Date"], dowjones["Price"])
plt.show()

# color ==> 'red', 'g', '#FF5733'
# linestyle ==> '-', '--', ':', '-.'
# marker ==> 'o', 's', '^'
# label ==> name for legend
```

## Scatter Plot

```python
tips=sns.load_dataset("tips")
tips.head()
```

```python
fig, ax=plt.subplots(figsize=(8,6))
ax.scatter(tips["total_bill"], tips["tip"])
plt.show()
```

## Histogram

```python
plt.figure(figsize=(8,6))
plt.hist(tips["total_bill"], bins=15)
plt.show()
```

## Pie

```python
titanic=sns.load_dataset("titanic")
titanic.head()
```

```python
plt.figure(figsize=(6,6))
plt.pie(x=titanic["sex"].value_counts(), labels=titanic['sex'].value_counts().index.str.capitalize(), autopct="%.1f%%", startangle=90)
plt.title('Passengers by Gender')
plt.show()
```

## Column Chart

```python
fig, ax= plt.subplots(figsize=(8,6))
ax.bar(titanic["class"].value_counts().index, titanic["class"].value_counts().values)
ax.set_title("Total Passenger by Class")
ax.set_xlabel("Class")
ax.set_ylabel("Total Passenger")
plt.show()
```

## Bar Chart

```python
total_passengers_by_town = titanic['embark_town'].value_counts().sort_values(ascending=True)
fig, ax= plt.subplots(figsize=(8,6))
ax.barh(total_passengers_by_town.index, total_passengers_by_town.values)
ax.set_title("Total Passenger by Town")
plt.show()
```

```python
# Create summary table
summary = titanic.groupby(['class', 'survived']).size().unstack(fill_value=0)

# Rename columns
summary.columns = ['Unservice', 'Service']

# Plot stacked column chart
summary.plot(
    kind='bar',
    stacked=True,
    figsize=(8, 5)
)

# Customize chart
plt.title('Total Passengers by Survival and Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Number of Passengers')
plt.legend(title='Status')
plt.xticks(rotation=0)

plt.tight_layout()
plt.show()
```
