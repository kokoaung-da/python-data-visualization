# Matplotlib Data Visualization

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sns.get_dataset_names()
```

```python
dowjones=sns.load_dataset("dowjones")
dowjones.head()
```

```python
tips=sns.load_dataset("tips")
tips.head()
```

## Line Plot

```python
fig, ax= plt.subplots(figsize=(8,6))
ax.plot(dowjones["Date"], dowjones["Price"])
plt.show()

# color ==> 'red', 'g', '#FF5733'
# linestyle ==> '-', '--', ':', '-.'
# marker ==> 'o', 's', '^'
# label ==> name for legend
```

## Bar Chart

```python
fig, ax= plt.subplots(figsize=(8,6))
ax.bar(tips["day"].value_counts().index, tips["day"].value_counts().values)
ax.set_title("Tips by Day")
ax.set_xlabel("Day")
ax.set_ylabel("Count")
plt.show()
```

```python
fig, ax=plt.subplots(figsize=(8,6))
ax.bar(tips["day"], tips["total_bill"], label="Total bill")
ax.bar(tips["day"], tips["tip"], label="Tips")
plt.show()
```

## Scatter Plot

```python
fig, ax=plt.subplots(figsize=(8,6))
ax.scatter(tips["total_bill"], tips["tip"])
plt.show()
```

## Histogram

```python
plt.figure(figsize=(8,6))
plt.hist(tips["total_bill"], bins=15)
plt.show()
```

## Pie

```python
plt.figure(figsize=(8,6))
plt.pie(x=tips["sex"].value_counts(), autopct="%.1f%%", labels=["Male", "Female"])
plt.show()
```
